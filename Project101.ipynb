```python
import cv2
from ultralytics import YOLO

# Load the pre-trained YOLOv8 model (pre-trained on COCO dataset)
model = YOLO('Car-Plate.pt')  # you can also try 'yolov8s.pt' for a more accurate model

# Access the laptop camera (0 refers to the default camera)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame")
        break

    # Run YOLO inference on the frame
    results = model(frame)

    # Plot the results on the frame
    annotated_frame = results[0].plot()  # The `results.plot()` method works here to draw boxes

    # Display the frame with detection
    cv2.imshow('YOLOv8 Live Person Detection', annotated_frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera and close the window
cap.release()
cv2.destroyAllWindows()
```
